## 4. Upload files and zip directly from the url

````
# Helper to upload directly file to your drive
import urllib.request
def download_file_from_url(url, file_name):
    rsp = urllib.request.urlopen(url)
    with open(file_name,'wb') as f:
        f.write(rsp.read())
````

````
download_file_from_url('https://images.unsplash.com/photo-1506973035872-a4ec16b8e8d9?ixlib=rb-1.2.1&q=80&fm=jpg&crop=entropy&cs=tinysrgb&dl=dan-freeman-7Zb7kUyQg1E-unsplash.jpg&w=640',
                       'sydney.jpg')
````

````
# Helper to upload directly zip file to your drive
import requests, zipfile, io

def download_zip_from_url(url, file_location):
    r = requests.get(url)
    zip_ref = zipfile.ZipFile(io.BytesIO(r.content))
    zip_ref.extractall(file_location)
````

## **Bonus :** hack scripts

1. To **stay connected to the Colab cloud machine** when you have a long script to run, open the dev tools and launch in the console:
```
function ClickConnect(){
    console.log("Working"); 
    document.querySelector("colab-toolbar-button").click() 
};
interval = setInterval(ClickConnect,300000);
```
This will create a section every 5 minutes, and prevent disconnection. To stop it:
```
clearInterval(interval)
```

2. To **get more RAM**, simply launch this:
```
a = []
while(1):
    a.append('1')
```  
Once your machine crashes, Google will raise a message asking you if you wanna switch to more RAM, just say yes and you will be rewarded with 25GB of RAM instead of 12GB.
